```python
def clean_and_escape_xml(input_file_name, output_file_name):
    with open(input_file_name, 'r', encoding='utf-8', errors='ignore') as infile, \
         open(output_file_name, 'w', encoding='utf-8') as outfile:
        line_num = 0
        for line in infile:
            line_num += 1
            try:
                line = line.replace("", "'") #Removing problem charecter 
                escaped_line = line.replace("&", "and") #Replacing & with and since it causes errors

                outfile.write(escaped_line)
            except Exception as e:
                print(f"Error processing line {line_num}: {e}")

```

```python
import xml.etree.ElementTree as ET
import csv
def parsing_files(input_file,positive_or_not,mode='a'):
  clean_and_escape_xml(input_file, 'reviews_cleaned.xml')
  output_file_name = 'every_reviews.csv'

  with open('reviews_cleaned.xml', 'r', encoding='utf-8') as file:
    data = file.read()
  data = "<reviews>" + data + "</reviews>"

  root = ET.fromstring(data)

  with open(output_file_name, mode, newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    if mode=='w':
      writer.writerow(['unique_id', 'asin', 'product_name', 'product_type', 'helpful', 'rating', 'title', 'date', 'reviewer', 'reviewer_location', 'review_text', 'positive_or_not'])

    for review in root.findall('review'):
      unique_id = review.find('unique_id').text.strip() if review.find('unique_id') is not None else ''
      asin = review.find('asin').text.strip() if review.find('asin') is not None else ''
      product_name = review.find('product_name').text.strip() if review.find('product_name') is not None else ''
      product_type = review.find('product_type').text.strip() if review.find('product_type') is not None else ''
      helpful = review.find('helpful').text.strip() if review.find('helpful') is not None else ''
      rating = review.find('rating').text.strip() if review.find('rating') is not None else ''
      title = review.find('title').text.strip() if review.find('title') is not None else ''
      date = review.find('date').text.strip() if review.find('date') is not None else ''
      reviewer = review.find('reviewer').text.strip() if review.find('reviewer') is not None else ''
      reviewer_location = review.find('reviewer_location').text.strip() if review.find('reviewer_location') is not None else ''
      review_text = review.find('review_text').text.strip() if review.find('review_text') is not None else ''

      writer.writerow([unique_id, asin, product_name, product_type, helpful, rating, title, date, reviewer, reviewer_location, review_text, positive_or_not])

  print("Successfully Added to CSV file")
```
